```python
import random
import matplotlib.pyplot as plt
import numpy as np
def roll_dice():
  return random.randint(1,6)

x=np.array([])
y=np.array([])

for z in range(10000):
  x=np.append(x,z)
  y=np.append(y,roll_dice()+roll_dice())

#print(x)
#print(y)
b=np.unique(y,return_counts=True)
##print(b)
plt.bar(b[0],b[1])
plt.show()
```
