```python
# simulate descreate randome variable
# let's take dice for example
# where we have outcome is fixed X -> {1,2,3,4,5,6}
import numpy as np
import matplotlib.pyplot as plt
import random as rd
n=1000
# we will role ${n} times
def roll_dice():
    return rd.randint(1,6)

indexs=np.array([])
results=np.array([])
for i in range(n):
  indexs=np.append(indexs,i)
  results=np.append(results,roll_dice())

plt.bar(indexs,results)
plt.show()
plt.scatter(indexs,results)
plt.show()
plt.plot(indexs,results)

# Here we have role the dice 100 times and found it's totally random and demostrate the 3 diffrent chart in numpy
```

```python
# Let's find out how many times each number is occures
from collections import Counter
counter=Counter(results)
new_values=[(i / n) for i in counter.values()]
plt.bar(counter.keys(),new_values)


# Conclusion we found that as we increase number of time role dice as given
# here value n then probility of each time comes for every single item nealry 1/6
# So we can say that probebility distribution of dice outcome is uniform.

# Dice is descrete uniform probabilitiy distribution.
```
