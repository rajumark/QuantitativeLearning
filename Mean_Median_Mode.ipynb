```python
from collections import Counter
marks=[45, 67, 89, 72, 56, 91, 38, 64, 77, 83,
         59, 70, 48, 95, 62, 38, 81, 53, 68, 86]


mean=sum(marks)/len(marks)
print("Mean="+str(mean))


marks.sort()
print(marks)
if(len(marks)%2==0):
  index=int(len(marks)/2)
  print("Median="+str(marks[index]))
else:
  index=int((len(marks)+1)/2)
  print("Median="+str(marks[index]))

counter=Counter(marks)
mode=counter.most_common(1)
print("Mode="+str(mode))
```
